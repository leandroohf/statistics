## **Distributions and Random Variables**

In probability theory, a **random variable** is a function that assigns a numerical value to each outcome in a sample space. A **distribution** describes how values of a random variable occur.

---

### **1. Discrete Distribution**

A **discrete random variable** takes on **countable** values (e.g., integers).

A random variable \( X \) is **discrete** if it has a probability mass function (PMF) $P(X = x)$, where:

$
P(X = x_i) \geq 0, \quad \sum_{i} P(X = x_i) = 1 \\ 
P(X = x_i) = p_i
$

The expected values:

$
E[X] = \sum_i x_iP(X = x_i)
$

You can define a discrete probability distribution in terms of f(x) (PDF) snd the $\delta(x)$ Dirac delta function.

$
f(x) = \sum_{i} p_i \delta(x - x_i) \\ 
\int_{-\infty}^{\infty} f(x) dx =  \int_{-\infty}^{\infty} \sum_{i} p_i \delta(x - x_i) dx = \sum_{i} p_i = 1 \\
E[X] = \int_{-\infty}^{\infty} x (\sum_{i} p_i \delta(x - x_i)) dx \\ 
E[X] = \int_{-\infty}^{\infty} (\sum_{i} x p_i \delta(x - x_i)) dx \\
E[X] = \sum_{i} \int_{-\infty}^{\infty} (x p_i \delta(x - x_i)) dx \\
E[X] = \sum_i x_i p_i
$

where we use the property of Dirac delta function:

$
\int_{-\infty}^{\infty} f(x) \delta(x - x_i) dx = f(x_i) \\
$

Examples:
- **Rolling a fair six-sided die**: \( X \) represents the outcome (1, 2, 3, 4, 5, or 6).  
  - PMF: $P(X = k) = \frac{1}{6}, \quad k = 1,2,3,4,5,6$.
- **Number of customers arriving at a bank** (modeled using a Poisson distribution).

---

### **2. Continuous Distribution**

A **continuous random variable** takes on an **uncountable** number of values, usually in an interval.

A random variable \( X \) is **continuous** if it has a probability density function (PDF) $f(x)$ such that:

$
P(a \leq X \leq b) = \int_{a}^{b} f(x) dx.
$

For any continuous random variable \( X \), its **expected value** is:

$
E[X] = \int_{-\infty}^{\infty} x f(x) dx.
$

Examples:
- **Height of people** in a population (modeled using a normal distribution).
- **Time between arrivals of buses** (modeled using an **exponential distribution**).

### **3. Joint Distribution**
A **joint distribution** describes the probability distribution of **two or more** random variables occurring together. It specifies how likely different combinations of values are.

---

#### **Mathematical Definition**
For two random variables \( X \) and \( Y \), the **joint probability distribution** is:

- **Discrete case (Joint Probability Mass Function, PMF)**:
  $
  P(X = x, Y = y) = f(x, y)
  $
  where:
  - \( P(X = x, Y = y) \) is the probability that \( X = x \) and \( Y = y \) occur **simultaneously**.
  - The sum over all possible values must be **1**:
    $
    \sum_{x} \sum_{y} P(X = x, Y = y) = 1.
    $
 
- **Continuous case (Joint Probability Density Function, PDF)**:
  $
  f(X, Y) = f(x, y)
  $
  where:
  - \( f(x, y) \) is the joint density function.
  - The total probability over all possible values must integrate to **1**:
    $
    \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f(x, y) dx dy = 1.
    $

    $
    E[X] = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} x f(x,y) dx dy. \\
    E[XY] = \int \int xy f(x,y) dxdy
    $

#### **Independent random variables**
 
If two random variables \(X\\) and \(Y\\) are independent, then their **joint probability distribution** can be factored as two separate functions:

$
f(X, Y) = f_1(X) f_2(Y)
$

This is the same of saying for probabilities:

$
P(X \in A \text{ and } Y \in B) = P(X)P(Y)
$

1. if knowing the value of one provides no information about the other.
2. This means their joint behavior is simply the product of their individual behaviors, implying no relationship between them.
  
  We going to show that $f_1(X)$ and $f_2(Y)$ are the **marginal distributions**. Lets started fromt the maeginal defition:
  
  $
  f_X(x) = \int f(x,y) dy \\
  f_X(x) = \int f_1(x) f_2(y) dy \\
  f_X(x) =  f_1(x) \int f_2(y) dy
  $
  
  Similar:
  
  $
 f_Y(y) =  f_2(y) \int f_1(x) dy
  $
  
  Because $\int \int f(x,y) dx dy = 1 $ one solution is force $ \int f_1(x) dx = 1$ and $ \int f_2(y) dy = 1$. Than we conclude:
  
  $
 f_X(x) =  f_1(x)\\
 f_Y(y) =  f_2(y)\\
  $
  
We claim that independence decomposes the joint distribution into the product of the marginal distributions. We say "independence allows decomposition".
  
The expectation of independent variable:

$

## **1. Expectation (E[X])**

The expectation (or expected value) of a random variable \(X\) is its long-run average value.

1. **INDEPENDENCE Rule**:
   If \( X \) and \( Y \) are independent:
   $
   E[XY] = E[X]E[Y]
   $
   
   **Proof**
   
   $
   E[XY]= \int \int xy f(x,y) dx dy \\
   = \int \int xy f(x)f(y) dx dy \\
   =  (\int x f(x) dx) *  (\int y f(y) dy) \\
   = E[X]E[Y]
   $

1. **Linearity of Expectation**:
   $
   E[aX + bY] = aE[X] + bE[Y]
   $
   
 **Linearity of expectation doesn’t care about independence** This is huge.  Even if two random variables are correlated or dependent, their expected values still add.
 
 ** E[D = X - Y]  and E[S= X+Y] for correlated random variables are uncorrelated. $Corr(D,S) = 0$ with Corr(X,Y) = \pho** 
   
1. **Expectation of a Constant**:
   $
   E[c] = c
   $

1. **Law of Iterated Expectations (LIE)**:
   $
   E[X] = E[E[X \mid Y]]
   $

## **2. Variance (Var)**

Variance measures the spread of a random variable.

### **Definition**
$
\text{Var}(X) = E[(X - E[X])^2] \\
 = E[(X-E[X])(X-E[X])] \\
 = E[X^2 -2XE[X] + E^2[X]] \\
 = E[X^2] - 2E[X]E[X] + E^2[X] \\
 = E[X^2] - (E[X])^2
$

### **Properties**
1. **Scaling Rule**:
   $
   \text{Var}(aX) = a^2 \text{Var}(X)
   $
2. **Variance of a Sum**:
   $
   \text{Var}(X + Y) = \text{Var}(X) + \text{Var}(Y) + 2\text{Cov}(X, Y)
   $
    * **Variance of Independent Variables**:
       If \( X \) and \( Y \) are independent:
       $
       \text{Var}(X + Y) = \text{Var}(X) + \text{Var}(Y)
       $

---

## **3. Covariance (Cov)**
Covariance measures the **linear relationship** between two random variables.

### **Definition**
$
\text{Cov}(X, Y) = E[(X - E[X])(Y - E[Y])] \\
= E[XY] - E[X]E[Y]
$

### **Properties**
1. **Symmetry**:
   $
   \text{Cov}(X, Y) = \text{Cov}(Y, X)
   $
2. **Scaling Rule**:
   $
   \text{Cov}(aX, bY) = ab \cdot \text{Cov}(X, Y)
   $
3. **Covariance of Independent Variables**:
   If \( X \) and \( Y \) are independent:
   $
   \text{Cov}(X, Y) = 0
   $

---

## **4. Correlation (Corr)**
Correlation normalizes covariance to a range of \([-1,1]\).

### **Definition**
$
\text{Corr}(X, Y) = \frac{\text{Cov}(X, Y)}{\sqrt{\text{Var}(X) \cdot \text{Var}(Y)}}
$

### **Key Points**
- If **Corr(X, Y) = 1** → Perfect positive linear relationship.
- If **Corr(X, Y) = -1** → Perfect negative linear relationship.
- If **Corr(X, Y) = 0** → No linear relationship (but can still be dependent).
- Can you ADD RELATION $R^2$ of the linear regression ?

### **Correlation and \( R^2 \) in Linear Regression**

In **simple linear regression** $(Y = a + bX + \varepsilon$), the **coefficient of determination** $R^2$ measures how well  $X$ explains the variance in $Y$. It is related to correlation as:

$
R^2 = \text{Corr}^2(X, Y).
$

#### **Key Insights:**
- $R^2$ is **always non-negative**, since squaring removes the sign of correlation.
- If $R^2 = 1$, \( X \) explains **100% of the variance** in \( Y \).
- If $R^2 = 0$, \( X \) explains **none** of the variance in \( Y \).

Thus, **$R^2$ measures the proportion of variance in \( Y \) that is explained by \( X \)**, while $\text{Corr}(X, Y)$ measures the strength and direction of the linear relations

## **5. Key Probability Distributions**
### **Normal Distribution**

$
X \sim \mathcal{N}(\mu, \sigma^2)
$
- **PDF**: $f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$
- **Expectation**: $E[X] = \mu$
- **Variance**: $\text{Var}(X) = \sigma^2$

Intuitions:

* normal distribution arises whenever many small random effects add up.
 Example: If you’re timing how long it takes you to get to work, and every little delay (traffic light, coffee spill, elevator wait) adds a small random component — your total commute time tends to be normal.
 
* It’s the “default” for many natural and human-made systems because of the Central Limit Theorem:

### **Multivariate Normal**

$
(X_1, X_2, ..., X_n) \sim \mathcal{N}(\mu, \Sigma) \\
(X,Y) \sim \mathcal{N}(\mu, \Sigma) \\ 
$

$
\mu =
\begin{bmatrix}
\mu_X \\
\mu_Y
\end{bmatrix}
$

$
\Sigma =
\begin{bmatrix}
\sigma_X^2 & \rho \sigma_X \sigma_Y \\
\rho \sigma_X \sigma_Y & \sigma_Y^2
\end{bmatrix}
$


where:
- $\mu_X = E[X]$ is the mean of \( X \).
- $\mu_Y = E[Y]$ is the mean of \( Y \).
- $\sigma_X^2 = \text{Var}(X)$ is the variance of \( X \).
- $\sigma_Y^2 = \text{Var}(Y)$ is the variance of \( Y \).
- $\rho = \text{Corr}(X,Y)$ is the correlation coefficient between \( X \) and \( Y \).
- The off-diagonal terms represent the **covariance**: $\text{Cov}(X, Y) = \rho \sigma_X \sigma_Y$.

**Linear combinations of normal variables are **also normal**.

### **Chi-Square Distribution**

If you have $Z_1, Z_2, …, Z_k \sim \mathcal{N}(0, 1)$ i.i.d., then the sum of their squares follows a 

$
X = \sum^k_{i=1} Z_i \\
Z_i \sim N(0,1) \\ 
X \sim \chi^2_k \\
$

Chi-squared distribution with k degrees of freedom:

$$
f(x) = \frac{1}{2^{k/2} \, \Gamma(k/2)} \, x^{k/2 - 1} e^{-x/2}, \quad x > 0
$$

- If $X_i \sim \mathcal{N}(0,1)$, then $\sum X_i^2 \sim \chi^2_k$

 **The sum of squared random variables that follows normal distribution follows the chi squared distribution**
 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# x-axis range
x = np.linspace(0, 20, 500)

# Degrees of freedom to plot
dfs = [1, 2, 5, 10]

# Plot Chi-squared PDFs
plt.figure(figsize=(10, 6))
for df in dfs:
    plt.plot(x, chi2.pdf(x, df), label=f'df = {df}')

plt.title("Chi-squared Distributions for Various Degrees of Freedom")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

### **Student’s t-Distribution**

$
t = \frac{z}{\sqrt{s^2/k}}, \quad z = x - E[X], z \sim \mathcal{N}(0,1), \quad S^2 \sim \frac{\sigma^2}{k}\chi^2_k
$

$$
f(t) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu\pi} \, \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}}
$$

* $s^2$ is the estimation of population variance using the sample variance
* E[X] is the population mean
* $\sigma^2$ : is population variance
* $\nu = k = n -1$: is the degree of freedom

Intuition

* Looks like a normal distribution but has heavier tails; depends on degrees of freedom.
* You use it when you don’t know the population variance and must estimate it from a small sample.
* The heavier tails reflect extra uncertainty. 
* As your sample size increases, your estimate of the variance gets better. the distribtion tends to look like a normal distribuiton.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# x-axis range for the plot
x = np.linspace(-5, 5, 500)

# Degrees of freedom for t-distribution
dfs = [1, 5, 30]  # You can adjust or add more if you want

# Plotting
plt.figure(figsize=(10, 6))

# Plot the normal distribution
plt.plot(x, norm.pdf(x), label='Normal Distribution', color='black', linewidth=2)

# Plot t-distributions with different degrees of freedom
for df in dfs:
    plt.plot(x, t.pdf(x, df), label=f't-distribution (df={df})')

# Add legend and titles
plt.title("Student's t-distribution Converging to Normal Distribution")
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

## **6. Important Theorems**

### **Law of Large Numbers (LLN)**
$
\frac{1}{n} \sum_{i=1}^{n} X_i \to E[X] \quad \text{as } n \to \infty
$

- The sample mean converges to the expected value.

### **Central Limit Theorem (CLT)** Cool
$
\frac{\sum X_i - n\mu}{\sqrt{n\sigma^2}} \to \mathcal{N}(0,1) \quad \text{as } n \to \infty
$
- The sum of i.i.d. (independent and indentically distributed) random variables **approaches normality**, regardless of their original distribution.

**The distribution of the sampled mean will be normal distributed regardless the population distribution (the shape of the distribution)**

Intuitions:

* If you average enough independent measurements, the average will be approximately normal, even if the original data is not.

* That is why the error distributions on Phisics experiments tends to be normal.

### **Jensen’s Inequality**

For a convex function \( f \):
$
E[f(X)] \geq f(E[X])
$

## **8. Monte Carlo Simulation (Basic Idea)**
Monte Carlo methods use random sampling to approximate mathematical results.

**Steps:**
1. Generate random variables from the given distribution.
2. Compute the function of interest.
3. Repeat many times and take the average.

Example: Estimating $\pi$ using random sampling in a unit square.
